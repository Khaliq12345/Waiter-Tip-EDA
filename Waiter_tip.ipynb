```python
import seaborn as sns
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
t = data['time'].value_counts()
sns.countplot(x = data['time'],data=data,hue=data['time'],palette = 'Set2')
plt.subplot(222)
sns.countplot(x = data['sex'], data=data, hue=data['sex'],palette = 'afmhot')
plt.subplot(223)
sns.countplot(x = data['smoker'], data=data, hue=data['smoker'], palette = 'Set1')
plt.subplot(224)
sns.countplot(x= data['day'], data=data, hue=data['day'], palette='gnuplot')
plt.tight_layout()
```

From the countplot we notice that most of the tip are given during launch time and most male customers give tip than female. It is also noticed that non-smokers give the tip most of the time and it is given mostly on saturdays.

```python
# statistical analysis of the numerical variables
data.describe()
```

From the statistical summary it is noticed that the average bill paid by customers is 19.8 and the average tip given is 3. We can as well conclude that the minimum and maximum bill paid is 3.07 and 50.81 respectively. 

```python
# histogram plot of the numerical variables
plt.figure(figsize = (10,5))
plt.subplot(131)
sns.histplot(data['total_bill'],element='bars',kde=True)
plt.subplot(132)
sns.histplot(data['tip'],color='g',element='bars',kde=True)
plt.subplot(133)
sns.histplot(data['size'],color='r',element='bars',kde=True)
plt.tight_layout()
```

From the visualisation above we can see that most of the bill lies between 17 - 19 and the tips paid lies between the range of 2-3.
From the histogram it is clear that on a table it is most likely to find 2 customers.

## ***Bivariate Analysis***


---
With the following analysis we will answer the following questions:


1.   The relationship between the total bill and the tip
2.   The relationship between the size of the table and the tip given 









```python
# plotting relationships between the dataset
plt.figure(figsize = (15,10))
plt.subplot(1,2,1)
sns.scatterplot(data=data,x=data['total_bill'],y=data['tip'],hue=data['sex'],palette='cividis_r',legend=False)
plt.subplot(122)
sns.scatterplot(data=data,x=data['size'],y=data['tip'],hue=data['sex'],palette='cividis_r',legend=False)
```

From th plot above we notice that there is a positive relationship between the total bill and the tip given. So as the total bill increases the tip given increases too.
But there is no relationship between the size and the people.

```python
#checking correlation
corr = data.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='YlGnBu')
```

From the above plot it is noticed that there is quite a strong correlation betweeen total_bill and tip and weak correlation between size and tip.
